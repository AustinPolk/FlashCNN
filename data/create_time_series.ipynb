```python
data = r"3993249.csv"
```

```python
import pandas as pd
import torch
import numpy as np

def create_tensors(using_stations):
    df = pd.read_csv(data)
    df = df[['STATION', 'DATE', 'PRCP']]
    df = df.pivot(index='DATE', columns='STATION', values='PRCP')
    df = df[using_stations]
    df = df.loc['1938-04-01':]
    df = df.ffill()

    daily = []
    observations = []

    for idx, row in df.iterrows():
        date = pd.to_datetime(row.name)
        obs  = torch.tensor(row.values)

        d = date.day_of_year
        dt = torch.tensor([d / 365.25, np.sin(2 * np.pi * d / 365.25), np.cos(2 * np.pi * d / 365.25)])

        daily.append(dt)
        observations.append(obs)

    without_days = torch.stack(observations, dim=0)
    print(without_days)

    days = torch.stack(daily, dim=0)
    print(days)

    # tensor of daily observations plus day of year info
    with_days = torch.cat((without_days, days), dim=1)
    print(with_days)

    stations_str = '-'.join(x for x in using_stations)
    torch.save(with_days, f'data-{stations_str}.bin')
```

```python
create_tensors(['USC00251825', 'USC00254335', 'USC00255810'])
create_tensors(['USC00254335', 'USC00255810', 'USW00014935'])
create_tensors(['USC00251825', 'USC00254335', 'USC00255810', 'USW00014935'])
```
